```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data.csv')
df
```

```python
df.head()
```

```python
df.drop_duplicates(inplace=True)
```

```python
print(df.isnull().sum())
```

```python
# Drop columns with more than 70% missing values
#threshold = 0.7 * len(df)
#df = df.dropna(thresh=threshold, axis=1)
print(df.columns)
print(df.isnull().sum())
```

```python
df['colors'].fillna('Unknown', inplace=True)
df['manufacturer'].fillna('Unknown', inplace=True)
df['reviews.username'].fillna('Anonymous', inplace=True)
```

```python
df['reviews.rating'].fillna(df['reviews.rating'].median(), inplace=True)
df['reviews.numHelpful'].fillna(df['reviews.numHelpful'].median(), inplace=True)

```

```python
# Drop columns with more than 70% missing values
threshold = 0.7 * len(df)
df = df.dropna(thresh=threshold, axis=1)

```

```python
df['reviews.date'] = pd.to_datetime(df['reviews.date'], format='%Y-%m-%dT%H:%M:%SZ', errors='coerce')
df['reviews.date'] = pd.to_datetime(df['reviews.date'])
df['reviews.date'].fillna(df['reviews.date'].mode()[0], inplace=True)  # Fill with most common date

df['reviews.title'].fillna('No title', inplace=True)  # Fill missing review titles

# Verify missing values are handled
print(df.isnull().sum())
```

```python
print(df.describe())  # Summary statistics for numeric columns
print(df['brand'].value_counts())  # Count of products by brand
print(df['reviews.rating'].value_counts())  # Distribution of ratings
```

```python
print("Top 5 Brands:")
print(df['brand'].value_counts().head())

print("\nLeast Common Brands:")
print(df['brand'].value_counts().tail())
```

```python
brand_ratings = df.groupby('brand')['reviews.rating'].mean().sort_values(ascending=False)
print(brand_ratings)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.histplot(df['reviews.rating'], bins=5, kde=True)
plt.xlabel("Ratings")
plt.ylabel("Count")
plt.title("Distribution of Review Ratings")
plt.show()
```

```python
plt.figure(figsize=(10,5))
df['brand'].value_counts().head(10).plot(kind='bar', color='skyblue')
plt.xlabel("Brand")
plt.ylabel("Number of Reviews")
plt.title("Top 10 Brands with Most Reviews")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,6))
df['reviews.date'].value_counts().sort_index().plot()
plt.xlabel("Date")
plt.ylabel("Number of Reviews")
plt.title("Reviews Over Time")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.boxplot(x=df['reviews.rating'], y=df['reviews.numHelpful'])
plt.xlabel("Rating")
plt.ylabel("Number of Helpful Votes")
plt.title("Helpful Votes vs Ratings")
plt.show()
```

```python
best_brand = df.groupby('brand')['reviews.rating'].mean().idxmax()
print(f"The best-rated brand is: {best_brand}")
```

```python
worst_brand = df.groupby('brand')['reviews.rating'].mean().idxmin()
print(f"The worst-rated brand is: {worst_brand}")
```
